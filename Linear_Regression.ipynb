# Linear Regression Single Variable

```python
from sklearn import linear_model as sk
import pandas as pd
import matplotlib
import matplotlib.pyplot as mt

print("Pandas:", pd.__version__)
print("Matplotlib:", matplotlib.__version__)

```

```python
dc = pd.read_csv("Book1.csv")
dc["pr"]
```

```python
gf = mt.scatter(dc.area,dc.price,marker="+",color="r")
mt.ylabel("price")
mt.xlabel("area")

gf
```

```python
rd = sk.LinearRegression()
rd.fit(dc[["area"]],dc.price)
```

```python
rd.predict([[1600]])
```

```python
dc["pr_price"] = rd.predict(
    dc[["pr"]].rename(columns={"pr": "area"})
)

dc
```
